```python
fig, ax = plt.subplots()
x = np.linspace(0,20,1000)
ax.plot(x, np.cos(x))
ax.axis('equal')
ax.annotate('local maximum', xy=(6.28,1),
           xytext=(10,4),
           arrowprops = dict(facecolor='black',
                            shrink=0.05))
ax.annotate('local minimum', xy=(5*np.pi,-1),
           xytext=(2,-6),
           arrowprops = dict(arrowstyle ="->",
                             connectionstyle="angle3,angleA=45, angleB=-110"));
plt.show()
```

```python
w = 8
h = 6
d = 70
plt.figure(figsize=(w,h),dpi=d)
plt.axis([0,5,0,5])
x = [0,3,5]
y = [1,4,3.5]
label_x = 1
label_y = 4
arrow_x = 3
arrow_y = 4

arrow_properties = dict(facecolor='black', width=1.5,headwidth=4, shrink=1.5)

plt.annotate("maximum", xy=(arrow_x, arrow_y), xytext=(label_x, label_y), arrowprops = arrow_properties)
plt.plot(x,y)
plt.show()
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection = '3d')
z = np.linspace(0,20,100)
x = z * np.tan(25 * z)
y = z * np.cos(25 * z)

ax.plot3D(x,y,z, 'green')
ax.set_title('3D')
```
